# 1. Write a program that takes coefficients A, B, C, D, and E as inputs representing a 4th degree polynomial in the form Ax^4 + Bx^3 + Cx^2 + Dx + E. Calculate the values of this polynomial for x in the range from -100 to 100, with constant discrete intervals.

Store the resulting x and y values as a NumPy array, where x represents the input values, and y represents the corresponding output values of the polynomial. Finally, use Matplotlib to plot the graph using the generated NumPy array.


```python
import numpy as np
import matplotlib.pyplot as plt

A, B, C, D, E = 1, -2, 3, 4, 5
x = np.arange(-100,101)
y = A * x**4 + B * x**3 + C * x**2 + D * x + E

#plotting graph size
plt.figure(figsize=(8, 6))
#plotting the graph and labeling the polynimal
plt.plot(x, y, label=f"{A}x^4 + {B}x^3 + {C}x^2 + {D}x + {E}")

plt.xlabel("x")
plt.ylabel("y")
plt.title("4th Degree Polynomial")
plt.grid(True)

#showing the labels
plt.legend()

plt.show()
```

# 2. Suppose you have a dictionary containing information about monthly sales for different products over a period of time. The dictionary has the following structure.

sales_data = {

    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],

    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'Apr'],

    'Sales': [100, 150, 200, 120, 180, 220, 90, 110, 130]

}

Write a Python script to convert this dictionary into a pandas DataFrame, calculate the total sales for each product over the entire period, and then create a bar plot using matplotlib to visualize the total sales for each product.



```python
import pandas as pd

sales_data = {
'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],
'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'Apr'],
'Sales': [100, 150, 200, 120, 180, 220, 90, 110, 130]
}

# Converting to DataFrame
df = pd.DataFrame(sales_data)
df
saleA=df[df['Product']=='A']['Sales'].sum()
print(saleA)
print(len(df.Product))

# for adding the sales price
def sales_sum(str1):
    sum1 = 0
    for i in range(len(df.Product)):
        if df.Product[i] == str1:
            sum1 += df.Sales[i]
    return sum1

print('The total sales for the product A over the entire period is: ',sales_sum('A'))
print('The total sales for the product B over the entire period is: ',sales_sum('B'))
print('The total sales for the product C over the entire period is: ',sales_sum('C'))

total_sales = {'A':sales_sum('A'),'B':sales_sum('B'),'C':sales_sum('C')}

# Create a bar plot
plt.figure(figsize=(8, 6))
plt.bar(total_sales.keys(), list(total_sales.values()), color='skyblue')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.title('Total Sales by Product in the bar graph')
plt.grid(axis='y')
plt.show()
```

# 3. Create visualizations for the following mathematical functions using Matplotlib:

Plot the following single-variable functions over the range 

[10,10], and include a title and labels for the axes:

(1) y = cos(x)

(2) y = e^x

(3) y = log(x), where x>0


Generate surface plots for these multi-variable functions over the range 

x=[10,10] and y=[10,10] , ensuring to add a title and labels for all axes:

(1) z = cos(sqrt(x^2+y^2)

(2) z = e^(-(x^2+y^2))

(3) z =  log(x^2+y^2) where x^2+y^2>0




```python
# x range
x = np.linspace(-10, 10, 1000)

# y values for each function
y_cos = np.cos(x)
y_exp = np.exp(x)
y_log = np.log(x[x > 0])  # as log(0) DNE

# Creating subplots
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# y = cos(x)
axs[0].plot(x, y_cos, label='y = cos(x)')
axs[0].set_title('Cosine Function')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

# y = e^x
axs[1].plot(x, y_exp, label='y = e^x', color='orange')
axs[1].set_title('Exponential Function')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')

# y = log(x)
axs[2].plot(x[x > 0], y_log, label='y = log(x)', color='green')
axs[2].set_title('Logarithmic Function (x > 0)')
axs[2].set_xlabel('x')
axs[2].set_ylabel('y')

# for legend
for ax in axs:
    ax.legend()

# Adjusting spacing between subplots
plt.tight_layout()

# Showing the plots
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Define the range for x and y
x = np.linspace(-10, 10, 200)
y = np.linspace(-10, 10, 200)
X, Y = np.meshgrid(x, y)

# z values for each function
# z = cos(sqrt(x^2 + y^2))
Z1 = np.cos(np.sqrt(X**2 + Y**2))

# z = e^(-(x^2 + y^2))
Z2 = np.exp(-(X**2 + Y**2))

# z = log(x^2 + y^2) where x^2 + y^2 > 0
Z3 = np.log((X**2 + Y**2),where = X**2 + Y**2 > 0)


# Create subplots
fig = plt.figure(figsize=(12, 8))

# z = cos(sqrt(x^2 + y^2))
ax1 = fig.add_subplot(231, projection='3d')
ax1.plot_surface(X, Y, Z1)
ax1.set_title('z = cos(sqrt(x^2 + y^2))')

# z = e^(-(x^2 + y^2))
ax2 = fig.add_subplot(232, projection='3d')
ax2.plot_surface(X, Y, Z2)
ax2.set_title('z = e^(-(x^2 + y^2))')

# z = log(x^2 + y^2) where x^2 + y^2 > 0
ax3 = fig.add_subplot(233, projection='3d')
ax3.plot_surface(X, Y, Z3)
ax3.set_title('z = log(x^2 + y^2)')

# Setting common labels
for ax in [ax1, ax2, ax3]:
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')

# plt.tight_layout()
plt.show()
```

# 4. For the function J(w) = w^2 + (54/w), implement the bracketing method (choose your own a, b, n).


```python
def J(w):
    return w**2 + 54/w

a = 0.0001
b = 10
n = 1000


dw = (b-a)/n
w1 = a
w2 = w1 + dw
w3 = w2 + dw
while(w3 <= b):
    if J(w1) >= J(w2) <= J(w3):
        print(f"The required interval where the minimum is: [{w1},{w2}]")
        break
    else:
        w1 = w2
        w2 = w3
        w3 = w3 + dw
else:
    print('DNE')
```
