# 5. WAP to plot a 3-d graph of the helical wave signal using the scatter method and normal line method. Plot them separately and specify legend.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Helical wave parameters
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

# scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='b', marker='o', label='Helical Wave (Scatter)')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.legend()

plt.show()
```

```python
# normal line plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='Helical Wave (Line)', color='r')

# Labeling
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.legend()

plt.show()
```

# 8.Take a N X M integer array matrix with space separated elements ( N = rows and M  = columns). Your task is to print the transpose and flatten results using numpy



```python
import numpy as np

#for matrix dimensions
N, M = map(int, input("Enter the dimensions N x M of the matrix: ").split())

#input from the user
print(f"Enter the elements of the {N}x{M} matrix rowwise(1st row then 2nd and so on..., separated by spaces:")
matrix_elements = [list(map(int, input().split())) for _ in range(N)]
matrix = np.array(matrix_elements)

transposed_matrix = np.transpose(matrix)

flattened_matrix = matrix.flatten()

print("Transposed matrix:")
print(transposed_matrix)
print("Flattened matrix:")
print(flattened_matrix)
```

# 9. WAP to capitalize a column of names in a Pandas Dataframe.

Eg : Input : {'Name': ['john', 'bODAY', 'aNa', 'Peter', 'nicky'], 'Education': ['masters', 'graduate', 'graduate', 'Masters', 'Graduate'], 'Age': [27, 23, 21, 23, 24]}

Output : {'Name': ['John', 'Boday', 'Ana', 'Peter', 'Nicky'], 'Education': ['masters', 'graduate', 'graduate', 'Masters', 'Graduate'], 'Age': [27, 23, 21, 23, 24]}



```python
import pandas as pd

data = {
    'Name': ['john', 'bODAY', 'aNa', 'Peter', 'nicky'],
    'Education': ['masters', 'graduate', 'graduate', 'Masters', 'Graduate'],
    'Age': [27, 23, 21, 23, 24]
}

df = pd.DataFrame(data)

# capitalize the 'Name' column
df['Name'] = df['Name'].str.capitalize()

print(df)
```

# 10. Use the central difference method to find the first and second order derivatives of the function. Use the following function for testing the result. And also verify the result manually (Write on paper and upload jpg). Refer to section 2.5.1 of “Optimization for Engineering Design: Algorithms and Examples” by KALYANMOY DEB, 2nd edition


f(x) = 3x**2 + 2x

```python
def f(x):
    return 3*x**2 + 2*x
def deriv1f(x,h):
    return (f(x+h)-f(x-h))/(2*h)
def deriv2f(x,h):
    return (f(x+h)-2*f(x)+f(x-h))/(h**2)

x = 3
h = 50
print('At x=1 the function value is: ',f(x))
print('At x=1 the 1st derivative of the function value is: ',deriv1f(x,h))
print('At x=1 the 2nd derivative of the function value is: ',deriv2f(x,h))
```
